```python
## To find highly correlated feature pairs
corr_matrix=df.drop(columns=['Bankrupt?']).corr().abs()
upper=corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape),k=1).astype(bool))
high_corr=[column for column in upper.columns if any(upper[column]>0.85)]
high_corr[:10]
```

```python
## Removing the highly correlated
df_reduced=df.drop(columns=high_corr)
df_reduced.shape
```

```python
## separating target and feature variables
x=df_reduced.drop('Bankrupt?',axis=1)
y=df_reduced['Bankrupt?']
print(x.shape)
print(y.shape)
```

```python
## we do Train-Test Split(since dataset is imbalanced,we use stratify)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# class distribution in train set
print(y_train.value_counts())
print(y_train.value_counts(normalize=True))
```

If 96% are non-bankrupt:

Model can say "0" always
And get 96% accuracy 

But it will fail to detect bankrupt companies.

Accuracy becomes misleading.

```python
## Applying SMOTE(only on Training Data)
from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=42)
x_train_sm,y_train_sm=smote.fit_resample(x_train,y_train)
print("Before SMOTE:",y_train.value_counts())
print("After SMOTE:",y_train_sm.value_counts())
```

What SMOTE did:

 It created synthetic bankrupt samples
 It did NOT duplicate rows
 It generated new artificial minority examples

Now model will learn patterns of bankruptcy properly.
Before SMOTE:

Model could predict 0 always → 96% accuracy
But it would completely ignore bankrupt companies.


```python
## Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train_sm)
x_test_scaled=scaler.transform(x_test)
```

```python
## Comparing different Models to predict which is best.
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

The following classification models were trained and evaluated:

- Logistic Regression
- K-Nearest Neighbors (KNN)
- Decision Tree
- Random Forest
- Gradient Boosting
- Neural Network (MLP)


```python
models={}
## Logistic Regression model
log_model=LogisticRegression(max_iter=1000)
log_model.fit(x_train_scaled,y_train_sm)
models["Logistic Regression"]=log_model.predict(x_test_scaled)

## KNN
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train_scaled,y_train_sm)
models["KNN"]=knn_model.predict(x_test_scaled)

## Decision Tree
dt_model=DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train_sm,y_train_sm)
models["Decision Tree"]=dt_model.predict(x_test)

## Random Forest
rf_model=RandomForestClassifier(random_state=42)
rf_model.fit(x_train_sm,y_train_sm)
models["Random Forest"]=rf_model.predict(x_test)

## Gradient Boosting
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(x_train_sm, y_train_sm)
models["Gradient Boosting"] = gb_model.predict(x_test)

# Neural Network (MLP)
mlp_model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42)
mlp_model.fit(x_train_scaled, y_train_sm)
models["Neural Network"] = mlp_model.predict(x_test_scaled)
```

##  Evaluation Metrics Used

Models were evaluated using:

- Accuracy
- Precision
- Recall
- F1-score
- Confusion Matrix

Since the dataset was highly imbalanced, **Recall and F1-score for the Bankrupt class (Class 1)** were considered the most important metrics.


```python
## evaluating the all models
for name,predictions in models.items():
    print(f"{name}")
    print("Accuracy:",accuracy_score(y_test,predictions))
    print("\nconfusion Matrix:")
    print(confusion_matrix(y_test,predictions))
    print("\nclassification Report:")
    print(classification_report(y_test,predictions))
```
